### 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
#send get request to the webpage server to get the source code of the page
url = requests.get("https://www.imdb.com/list/ls091520106/")

#see page content
page = BeautifulSoup(url.content)

# top Movies name
name = page.find_all("h3", class_="lister-item-header")

# get text from movie name web elements
movies_name = []            #empty list
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text)


# Year of release
year = page.find_all("span",class_="lister-item-year text-muted unbold")
year_of_release = []        #empty list
for k in year:
    year_of_release.append(k.text)

      
# IMDB Rating
rating = page.find_all("div",class_="ipl-rating-star small")

# scrape text from rating web element
IMDB_rating = []          #empty list
for i in rating:
      IMDB_rating.append(float(i.text))
        
# Make data frame of top 100 movies on IMDB
IMDB_top_100=pd.DataFrame({})
IMDB_top_100['movies_name']=movies_name
IMDB_top_100['year_of_release']=year_of_release
IMDB_top_100['IMDB_rating']=IMDB_rating  
IMDB_top_100
```

### 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release).

```python
#send get request to the webpage server to get the source code of the page

url = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")

# check the page content
page = BeautifulSoup(url.content)

# top Movies name
name = page.find_all("td", class_="titleColumn")

# get text from movie name web elements
movies_name = []            #empty list
for i in name:
        movies_name.append(i.text.replace("\n",""))


# Year of release
year = page.find_all("span",class_="secondaryInfo")
year_of_release = []        #empty list
for k in year:
    year_of_release.append(k.text.strip())


# IMDB Rating
rating = page.find_all("td",class_="ratingColumn imdbRating")

# scrape text from rating web element

IMDB_rating = []            #empty list
for i in rating:
      IMDB_rating.append(float(i.text.strip()))
        
# Make data frame of top 100 India imdb movies
indian_top_100=pd.DataFrame({})
indian_top_100['movies_name']=movies_name[:100]
indian_top_100['year_of_release']=year_of_release[:100]
indian_top_100['IMDB_rating']=IMDB_rating[:100]
indian_top_100    
```

### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
#send get request to the webpage server to get the source code of the page

page = requests.get('https://bookpage.com/reviews/')
soup = BeautifulSoup(page.content, 'html.parser')
url_tags = soup.find_all("h4",class_="italic")       # Extract books URL

urls = []
for i in url_tags:
    for j in i.find_all("a", href=True):
        if j.text:
            urls.append(j['href'])
book_name = []                                        #empty list
author_name = []                                      #empty list
book_genre = []                                       #empty list
books_review = []                                     #empty list

for url in urls:
        book = requests.get('https://www.bookpage.com'+url)
        soup = BeautifulSoup(book.content, 'html.parser')
        book_name.append(soup.find('h1').text.replace('\n','').replace('★',''))
        author_name.append(soup.find('h4').text.replace('\n',''))
        book_genre.append(soup.find('p', class_="genre-links").text.replace('\n',''))
        books_review.append(soup.find('div', class_= "article-body").text.replace('\n',''))
        

# Make data frame 
books=pd.DataFrame({})
books["Book Name"]=book_name[:5]
books["Author"]=author_name[:5]
books["Genre"]=book_genre[:5]
books["Review"]=books_review[:5]
books
```

### 5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

iii) Top 10 ODI bowlers along with the records of their team and rating.

### i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
#send get request to the webpage server to get the source code of the page
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page = requests.get(url)

# see content in page
soup = BeautifulSoup(page.content)

#scrape team names
team = soup.find_all("span",class_='u-hide-phablet')

team_name = []
for i in team:
    team_name.append(i.text)
matches = []                #empty list
points = []                 #empty list
ratings = []                #empty list
new_list = []               #empty list

for i in soup.find_all("td",class_='rankings-block__banner--matches'):    # first place team number of matches
    matches.append(i.text)
    
for i in soup.find_all("td",class_='rankings-block__banner--points'):     # first place team points
    points.append(i.text)
    
for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):# first place team ratings
    ratings.append(i.text.replace("\n",""))
    
for i in soup.find_all("td",class_='table-body__cell u-center-text'):  # other teams number of matches and points
    new_list.append(i.text)
    
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i])         # other teams matches
    points.append(new_list[i+1])        # other teams points
    
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):# other teams number of matches and ratings
    ratings.append(i.text)
    
# Make data frame of top 10 ICC teams
icc=pd.DataFrame({})
icc['Team_name']=team_name[:10]
icc['Matches']=matches[:10]
icc['Points']=points[:10]
icc['Ratings']=ratings[:10]
icc    
```

### ii) Top 10 ODI Batsmen in men along with the records of their team and rating. 

```python
#send get request to the webpage server to get the source code of the page
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
page = requests.get(url)

# see content in page
soup = BeautifulSoup(page.content)

players = []    #empty list
team_name = []  #empty list
rating = []     #empty list

for i in soup.find_all("div",class_='rankings-block__banner--name-large'):   # first place player name
    players.append(i.text)
for i in soup.find_all("div",class_='rankings-block__banner--nationality'):  # first place player team name
    team_name.append(i.text.replace("\n",""))
for i in soup.find_all("div",class_='rankings-block__banner--rating'):       # first place player rating
    rating.append(i.text)
for i in soup.find_all("td",class_='table-body__cell rankings-table__name name'):# players name
    for j in i.find_all('a'):
        players.append(j.text)
for i in soup.find_all("span",class_='table-body__logo-text'):    # players team name
    team_name.append(i.text)
for i in soup.find_all("td",class_='table-body__cell rating'):    # players rating
    rating.append(i.text)
    
# Make data frame of top 10 ICC Batsmen
Batsmen=pd.DataFrame({})
Batsmen['Player']=players[:10]
Batsmen['Team']=team_name[:10]
Batsmen['Rating']=rating[:10]
Batsmen
```

### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
#send get request to the webpage server to get the source code of the page
url = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YVRY9JpBzIU")


# see content in url
soup = BeautifulSoup(url.content)

period = []                 #empty list
short_desc = []             #empty list
temperature = []            #empty list
description = []            #empty list

for i in soup.find_all("div", class_="col-sm-2 forecast-label"):
    period.append(i.text)
for i in soup.find_all('div',class_="col-sm-10 forecast-text"):        
        short_desc.append(i.text.split(',')[0])
for i in soup.find_all('div',class_="col-sm-10 forecast-text"):        
        description.append(i.text)
        
for i in soup.find_all('p',attrs={'short-desc'}):
        if i.next_sibling is not None:
            temperature.append(i.next_sibling.text)
        else:
            temperature.append(' ')
for i in soup.find_all("div", class_="col-sm-10 forecast-text"):
        description.append(i.text)
        
# Make data frame
san_francisco_weather=pd.DataFrame({})
san_francisco_weather['Period']=period[:9]
san_francisco_weather['Temperature']=temperature[:9]
san_francisco_weather['Short_description']=short_desc[:9]
san_francisco_weather['Description']=description[:9]
san_francisco_weather
```

### 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
#send get request to the webpage server to get the source code of the page
url = requests.get("https://internshala.com/fresher-jobs")


# see content in url
soup = BeautifulSoup(url.content)

job_titles = []                  #empty list
companies_name = []              #empty list
CTC = []                         #empty list
apply_date = []                  #empty list


for i in soup.find_all('div',class_="heading_4_5 profile"):
    job_titles.append(i.text.strip())
for i in soup.find_all('a', class_="link_display_like_text"):
    companies_name.append(i.text.strip())
for i in soup.find_all('div', class_="item_body"):
    CTC.append(i.text.strip())
for i in soup.find_all('div', class_='item_body'):
    apply_date.append(i.text.strip())


jobs=pd.DataFrame({})
jobs['job_titles']=job_titles [:40]
jobs['companies_name']=companies_name [:40]
jobs['CTC']=CTC [:40]
jobs['apply_date']=apply_date [:40]
jobs
```
