### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
# Connect to the webdriver

driver = webdriver.Chrome("chromedriver.exe")
driver.get('https://www.naukri.com/')

# finding element for job search bar
    
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')
job_loc = driver.find_element_by_id('qsb-location-sugg')
job_loc.send_keys('Bangalore')
search_btn = driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
time.sleep(4)

# Create the required 4 empty lists to store the data while scraping

job_title=[]
job_location=[]
company_name=[]
experience_required=[]

# Exact all the tags having the job titles
    
title_tags = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
title_tags[0:10]  

# Exact the text from the above tags one by one by looping over these tags

for i in title_tags:
    title=i.text
    job_title.append(title)
job_title[0:10]

# Exact all the tags having the job location
    
loc_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
loc_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in loc_tags:
    location=i.text
    job_location.append(location)
job_location[0:10]

 # Exact all the tags having the companies name
    
companies_tags = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in companies_tags:
    company=i.text
    company_name.append(company)
company_name[0:10]

# Exact all the tags having the experience
    
exp_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
exp_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in exp_tags:
    experience=i.text
    experience_required.append(experience)
experience_required[0:10]


print(len(job_title))
print(len(job_location))
print(len(company_name))
print(len(experience_required))

 # creating the dataframe from the scraped data and taking only first 10 jobs
df=pd.DataFrame({"job_title":job_title[0:10],"job_location":job_location[0:10],"company_name":company_name[0:10],"experience_required":experience_required[0:10]})
df
```

### Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
# Connect to the webdriver

driver = webdriver.Chrome("chromedriver.exe")
driver.get('https://www.naukri.com/')

# finding element for job search bar
    
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Scientist')
job_loc = driver.find_element_by_id('qsb-location-sugg')
job_loc.send_keys('Bangalore')
search_btn = driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
time.sleep(4)
```

```python
# Create the required 4 empty lists to store the data while scraping

job_titles=[]
company_names=[]
locations_list=[]
full_job_description=[]

# Exact all the tags having the job titles
    
titles_tags = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in titles_tags:
    title=i.text
    job_titles.append(title)
job_titles[0:10]

# Exact all the tags having the companies name
    
companies_tags = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in companies_tags:
    company=i.text
    company_names.append(company)
company_names[0:10]

# Exact all the tags having the job location
    
loc_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
loc_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in loc_tags:
    location=i.text
    locations_list.append(location)
locations_list[0:10]
```

```python
# scraping the full job-description, for scraping full job description we have to go in each of the jobs separately

urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")]
for i in urls[0:10]:
    driver.get(i)
    time.sleep(3)
    
try:
    job_desc = driver.find_element_by_xpath("//section[@class='job-desc']")
    full_job_description.append(job_desc.text)
        
except NoSuchElementException :
        
    full_job_description.append("-")

# creating the dataframe from the scraped data and taking only first 10 jobs

print(len(job_titles))
print(len(company_names))
print(len(locations_list))
print(len(full_job_description))

df=pd.DataFrame({"job_titles":job_titles[0:10],"company_names":company_names[0:10],"locations_list":locations_list[0:10],"full_job_description":full_job_description[0:10]})
df
```

```python
# finding element for job search
search_job = driver.find_element_by_id("sc.keyword")
search_job.send_keys('Data Scientist')
```

```python
# finding element for job location
job_loc = driver.find_element_by_id("sc.location")
job_loc.send_keys('Noida')
job_loc.send_keys(Keys.ENTER)
```

```python
# Create the required 3 empty lists to store the data while scraping

company_name=[]
job_posted=[]
rating_company=[]

# Exact all the tags having the companies name
    
companies_tags = driver.find_elements_by_xpath("//div[@class='d-flex justify-content-between align-items-start']/a")
companies_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in companies_tags:
    company=i.text
    company_name.append(company)
company_name[0:10]
    
```

```python
# Exact all the tags having the rating of the company
    
rating_tags = driver.find_elements_by_xpath("//span[@class=' job-search-key-srfzj0 e1cjmv6j0']")
rating_tags[0:10]

# Exact the text from the above tags one by one by looping over these tags

for i in rating_tags:
    rating=i.text
    rating_company.append(rating)
rating_company[0:10]


# creating the dataframe from the scraped data and taking only first 10 jobs

print(len(company_name))
print(len(job_posted))
print(len(rating_company))


df=pd.DataFrame({"company_name":company_name[0:10],"job_posted":job_posted[0:10],"rating_company":rating_company[0:10]})
df
```
